###Creating tensors with tf.variable

Tensors created using tf.constant cannot be modified. To create mutable tensors, use tf.Variable

```python
#Creating tensor using tf.Variable
mutable_tensor = tf.Variable([10, 7])
mutable_tensor
```

```python
#Changing a value in the mutable_tensor
print(mutable_tensor[0]) #Prints 10
# mutable_tensor[0] = 100 # Causes an error. 
```

**Note:**
1. To change the value at an index, use .assign method
2. The .assign method cannot be used on a tensor created using tf.constant. It throws error

```python
mutable_tensor[0].assign(100)
mutable_tensor
```

### Creating random tensors

```python
random1 = tf.random.Generator.from_seed(7) #Seed gives us reproducability
random1 = random1.normal(shape=(3,2)) #Random values from normal distribution
random2 = tf.random.Generator.from_seed(7)
random2 = random2.normal(shape=(3,2)) #Generating from the same seed gives same tensor

random1, random2, random1 == random2
```

```python
random3 = tf.random.Generator.from_seed(7)
random3 = random3.uniform(shape=(3,2)) #Random values from uniform distribution
random3
```

### Shuffling order of elements in a tensor

```python
not_shuffled = tf.constant([[10, 7], [2, 5], [3, 4]])
tf.random.shuffle(not_shuffled, seed=42) #Does not affect the original tensor
```

**Important note about seed**:
Seed is of 2 types: global seed and operation-level seed. These affect the reproducibility of the operations of tf.random. Please read the documentation to find how setting seed affects the operation of tensorflow methods.
https://www.tensorflow.org/api_docs/python/tf/random/set_seed

### Other ways to create tensors

```python
tf.ones([10, 7]) #Creates a tensor with all ones with shape of (10, 7)
```

```python
tf.zeros([10, 7]) #Creates a tensor with all zeros of shape (10, 7)
```

```python
#Creating tensors from numpy array
import numpy as np
numpy_A = np.arange(1, 25) #Creates array of size 24
A = tf.constant(numpy_A) #Creates a tensor with shape (24, )
A
```

**Note:** We can create a tensor from numpy array and provide a specific shape. We need to ensure that the number of elements in the tensor based on the given shape must match the number of elements in the numpy array. In the above example, we have 24 elements in the numpy array. So the shape of the tensor can be (2, 12), (3, 8), (2, 3, 4) and even more with higher dimensions. We cannot form a tensor with shape (4, 7) because 24 != 28

```python
B = tf.constant(numpy_A, shape=(3, 8))
C = tf.constant(numpy_A, shape=(3, 2, 4))

# D = tf.constant(numpy_A, shape=(3, 9)) #NOT POSSIBLE

B, C
```

### Attributes of a tensor

Attributes of a tensor are: dtype, rank(number of dimensions), shape, size

```python
t = tf.zeros([2, 3, 4, 5])
print("Data type of the elements in the tensor is", t.dtype)
print("Rank (number of dimensions) of the tensor is", t.ndim)
print("Shape of the tensor is", t.shape)
print("Size (total number of elements in the tensor) is", tf.size(t).numpy())
```

### Indexing and expanding a tensor

Indexing a tensor is same as indexing a Python list. Negative indexing is also applicable

```python
#Adding a new dimension
#Method 1: Using tf.newaxis
t = tf.zeros([2, 3, 4, 5])
t_expanded = t[..., tf.newaxis] #Here ... means all the previous dimensions are copied
t_expanded.shape
```

```python
#Method 2: Using tf.expand_dims
t_Expanded = tf.expand_dims(t, axis=-1) #axis can be anything in [0, D-1] where D is current dimensions. Negative indexing is also possible
t_Expanded.shape
```

```python
#Reshaping a tensor
tensor = tf.constant([[10, 7, 8], [1, 2, 3]])
print(tensor.shape)
print(tf.reshape(tensor, shape=(3, 2))) #Does not change the original tensor
print(tensor.shape) #Shape is still (2, 3)
```

### Operations on tensors

#### Basic operations

```python
t = tf.constant([[10, 9], [4, 0]])
print(t + 10 )
print(t - 10 )
print(t * 10 )
print(t / 10 )
print(t / 0)
print(t)
#Adds/Substracts/Multiplie or divides with 10 for each element of the tensor
#These operations do not modify the original tensor
```

**Note:** Division with zero does not throw an error. We get "inf" or "nan"
For non-zero elements, we get "inf" meaning infinity
For zero, we get "nan" meaning not a number

**Note:** In tensorflow libray, we have math library which contains many functions like add, multiply, etc. The results look similar like using the regular operators of mathematical operations but the functions in math library provide higher speeds and it can be seen when used with large tensors. Documentation: https://www.tensorflow.org/api_docs/python/tf/math

#### Matrix multiplication

In Python, we have '@' operator to perform matrix multiplication.
If we use '*' operator, we get element wise multiplication.
In tensorflow, we can use tf.matmul function.

```python
X = tf.constant([[10, 7], [3, 4]])
Y = tf.constant([[2, 5], [8, 4]])
print(X @ Y)
print(tf.matmul(X, Y))
```

```python
#Transpose of a matrix
X = tf.constant([[10, 7], [3, 4]])
print(tf.transpose(X))
print(X) #Original remains same
```

#### Changing the data type of a tensor

```python
A = tf.constant([[10, 7], [3, 5]], dtype='float64')
B = tf.cast(A, dtype='float16')
print(A)
print(B)
```

#### Aggregating tensors

```python
#Finding min, max, mean, sum and more
E = tf.constant(np.random.randint(1, 100, size=(5, 5)))
print(E)
print(tf.reduce_min(E))
print(tf.reduce_max(E))
print(tf.reduce_sum(E))
print(tf.reduce_mean(E))
```

**Note**: Variance and standard deviation are not straightforward to find.
tf.reduce_std or tf.reduce_var are not available.

Variance: 
We can find variance using tf.math.reduce_variance function. The input tensor to this function must be of float type and not int. The above tensor E is of int type. Hence convert to float.

Standard deviation:
tf.math.reduce_std function can be used with input as float

```python
#Variance 
tf.math.reduce_variance(tf.cast(E, dtype='float32'))
```

```python
#Standard deviation
tf.math.reduce_std(tf.cast(E, dtype='float32'))
```

#### Positional max and min

```python
F = tf.random.normal([50])
print(F)

print(tf.argmax(F)) #Returns index of the largest value
print(F[tf.argmax(F)])
      
print(tf.argmin(F)) #Returns index of the minimum value
print(F[tf.argmin(F)])
```

```python
G = tf.random.uniform(shape=[1, 1, 1, 1, 1, 5])
print(G)

G_squeezed = tf.squeeze(G)
print(G_squeezed)
```
